# 9 Normalization

```python
max_hour = df['hours-per-week'].max()
min_hour = df['hours-per-week'].min()
scale = max_hour - min_hour
print('max_hour: ',max_hour)
print('min_hour: ',min_hour)
print('scale: ',scale)
```

```python
df_maxmin = df
df_maxmin['hours_max_min'] = (df_maxmin['hours-per-week'] - min_hour)/scale
df_maxmin.head(10)
```

```python
df_num = df[['age','education','hours-per-week']]
df_num.head(10)
```

```python
df_num.describe()
```

```python
def zscore(cols):
    res = []
    for col in cols:
        score = [round((e - col.mean())/col.std(),4) for e in col]
        res.append(score)
    return res
```

```python
df_zscore = df_num
```

```python
df_zscore[['age_zscore','edu_zscore','hours_zscore']] = zscore(df_zscore.values)
```

```python
df_zscore.head(10)
```
